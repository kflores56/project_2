### Adding dependencies for the extract, transform and load processes.

```python
#Dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

### The resources includes three CSVs files namely, Future50, Independence100 and Top250. These are loaded into the dataframes using pandas.

```python
#Extracting data from csv files
asylum_seekers_monthly =pd.read_csv("Resources/asylum_seekers_monthly.csv")
asylum_seekers_monthly.sample(20)
asylum_seekers_monthly = asylum_seekers_monthly.rename(columns ={"Country / territory of asylum/residence":"host_country", 
                                                                "Origin":"origin", "Year":"year", "Month":"month", "Value":"value"})
asylum_seekers_monthly
```

### Data clean up to remove the special characters and null values in all the three datasets

```python
#Replacing the special characters with zero.
asylum_seekers = asylum_seekers_monthly.copy()
asylum_seekers=asylum_seekers.replace(['*'], [0])
asylum_seekers.sample(20)
```

```python
#Retrieving the column names for reference
asylum_seekers.columns
```

```python
#Reading the second dataset using pandas
demographics = pd.read_csv("Resources/demographics.csv")
demographics.sample(20)
```

```python
#Replacing the special characters and null values with zero.
demographics = demographics.fillna(0)
demographics=demographics.replace(['*'], [0])
demographics=demographics.replace([''], [0])
demographics.sample(20)
```

```python
#Retrieving the column names for reference
demographics.columns
```

```python
#Renaming the column names in the standard format that is required in the database
labels= ['year', 'host_country', 'location_name',
       'female_0_4', 'female_5_11', 'female_5_17', 'female_12_17',
       'female_18_59', 'female_60plus', 'f_unknown', 'f_total', 'male_0_4',
       'male_5_11', 'male_5_17', 'male_12_17', 'male_18_59', 'male_60plus',
       'm_unknown', 'm_total']
demographics.columns = labels
demographics
```

```python
#Reading the third dataset using pandas
time_series = pd.read_csv("Resources/time_series.csv")
time_series.sample(20)
```

```python
#Retrieving the column names for reference
time_series.columns
```

```python
#Renaming the column names in the standard format that is required in the database
labels2= ['year', 'host_country', 'origin','population_type', 'value']
time_series.columns = labels2
#Replacing the special characters
time_series=time_series.replace(['*'], [0])
time_series
```

### Create connection to the database.

```python
#Creating the postgres connection
connection_string = "postgres:pgadmin@localhost:5432/project"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

### Load the final DataFrame into the database.

```python
#The data from the final dataframe is loaded to the database
# asylum_seekers.to_sql(name='asylumseekers', con=engine, if_exists='append', index =True)
# demographics.to_sql(name='demographics', con=engine, if_exists='append', index =True)
time_series.to_sql(name='time_series', con=engine, if_exists='append', index =True)
```
